```python
def pandas_mat_2_stack(dataframe):
    stack = dataframe.stack()
    stack1 = stack[stack >= 1].rename_axis(('source', 'target')).reset_index(name='weight')
    stack2 = stack[stack == 0].rename_axis(('source', 'target')).reset_index(name='weight')
    all_stack = pd.concat([stack1,stack2])
    return all_stack
    
```

```python
def stack_2_graph(stack):
    graph = nx.from_pandas_edgelist(stack,source="source",target="target",edge_attr=True)
    return graph
```

```python
stack2 = pandas_mat_2_stack(pr_dr)
```

```python
stw2 = stack_2_graph(stack2)
```

```python
stack2
```

```python
def node_2_vec(graph,save_name):
    node2vec = Node2Vec(graph,dimensions = 16,walk_length=10,num_walks = 10)
    model = node2vec.fit(window=10, min_count=1, batch_words=4)  
    model.save(save_name)
```

```python
def returnSimilars(model,value):
    return model.wv.most_similar(value)
```

```python
node_2_vec(stw2,"./pr_dr.model")
```
